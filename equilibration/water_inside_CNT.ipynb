## Number of water molecules inside the nanotubes as a function of time

```python
#Import MDAnalysis
import MDAnalysis as mda
import matplotlib.pyplot as plt
```

```python
#Load trajectory
simulacio = mda.Universe('../input/system.psf','MDequil.dcd')
```

```python
#init list
time =[]
waterin=[]
```

```python
#Loop over frame calculating water inside as a function of time
for ts in simulacio.trajectory:
    time.append(simulacio.trajectory.time)
    waterin.append(len(simulacio.select_atoms('type OT and prop abs z <6.')))
```

```python
#Plot
plt.plot(time,waterin)
plt.xlabel('Time (ps)')
plt.ylabel('Number of Water Molecules inside a Nanotube')
plt.show()
```
